# Lesson 1 ? First Harborlight Response

Learn the current Python SDK path: OpenAI(), client.responses.create(...), instructions, input, response.output_text, metadata, and usage.

**All Harborlight records are fictional.** Demo mode uses a recorded output fixture and makes no API call. Set HARBORLIGHT_LIVE=1 only when you intentionally want a billed live request.

## Input, deterministic data, and generated output

The packaged policy record is deterministic input. A live model response is generated and may vary. A recorded fixture is neither current nor live; its label is printed beside it.

The SDK's output_text convenience property is the ordinary way to read text. Do not assume text lives at response.output[0].content[0].text, because output can also contain reasoning or tool items.

```python
import os
from importlib.metadata import version

from openai import OpenAI

from harborlight_responses.config import api_key_available, resolve_model
from harborlight_responses.fixtures import FIRST_RESPONSE_FIXTURE, FIXTURE_LABEL
from harborlight_responses.responses import create_first_response, response_metadata
from harborlight_responses.services import get_policy

selection = resolve_model()
record = get_policy("FIC-HLA-1002")
live_requested = os.getenv("HARBORLIGHT_LIVE") == "1"
live = live_requested and api_key_available()

print("OpenAI SDK:", version("openai"))
print("Selected model:", selection.model)
print("Mode:", "Live API" if live else "Demo Fixture")
print("Deterministic input:", record.model_dump(mode="json"))

if live_requested and not live:
    print("Live mode was requested, but OPENAI_API_KEY is unavailable; no API call was made.")
```

```python
if live:
    client = OpenAI()
    response = create_first_response(client, model=selection.model, record=record)
    print()
    print("Generated output:")
    print(response.output_text)
    print()
    print("Response metadata and usage:")
    print(response_metadata(response))
else:
    print()
    print(FIXTURE_LABEL)
    print()
    print("Recorded generated output:")
    print(FIRST_RESPONSE_FIXTURE)
    print()
    print("Metadata: fixture mode has no live response object or billed usage.")
```

## Takeaway

Instructions define high-priority behavior for this request; input carries the fictional record. output_text aggregates generated text. Usage fields may be absent in fixtures and should be checked before display. No response identifier belongs in source control.
